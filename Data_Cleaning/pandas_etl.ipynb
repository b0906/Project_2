# Extract

### Extract CSVs into DataFrames

We start by loading the 2 csv files into dataframes and do .head() to see what's in the dataframes

```python
women_shoe_file = "../Resources/7003_1.csv"
women_shoe_df = pd.read_csv(women_shoe_file)
women_shoe_df.head()
```

```python
men_shoe_file = "../Resources/7004_1.csv"
men_shoe_df = pd.read_csv(men_shoe_file)
men_shoe_df.head()
```

### Here we create a new category column and set all values in the column to Women

```python
women_shoe_df["category"] = "Women"
women_shoe_df.head()
```

```python
men_shoe_df["category"] = "Men"
men_shoe_df.head()
```

# Transform

For the transform process, we selected the columns we needed, renamed them, and drop duplicates, set the index to the id column. Then, we merged the women's and men's dataframes together. After that, we split the merged dataframe into 2 dataframes called brand_df and shoe_data_df.

### Transform women's shoes DataFrame

```python
# Create a filtered dataframe from specific columns
women_shoe_cols = ["id", "brand", "category", "prices_amountmin", "prices_amountmax"]
women_shoe_transformed = women_shoe_df[women_shoe_cols].copy()

# Rename the column headers
women_shoe_transformed = women_shoe_transformed.rename(columns={"prices_amountmin": "prices_min",
                                                          "prices_amountmax": "prices_max",
                                                               "brand": "brand_name"})

# Clean the data by dropping duplicates and setting the index
women_shoe_transformed.drop_duplicates("id", inplace=True)
women_shoe_transformed.set_index("id", inplace=True)

women_shoe_transformed.head()
```

### Transform men's shoes DataFrame

```python
# Create a filtered dataframe from specific columns
men_shoe_cols = ["id", "brand", "category", "prices_amountmin", "prices_amountmax"]
men_shoe_transformed = men_shoe_df[men_shoe_cols].copy()

# Rename the column headers
men_shoe_transformed = men_shoe_transformed.rename(columns={"prices_amountmin": "prices_min",
                                                          "prices_amountmax": "prices_max",
                                                            "brand" : "brand_name"})
                                                            

# Clean the data by dropping duplicates and setting the index
men_shoe_transformed.drop_duplicates("id", inplace=True)
men_shoe_transformed.set_index("id", inplace=True)

men_shoe_transformed.head()
```

```python
# Concat the men's and women's shoes dataframes together
shoe_data_complete_df = pd.concat([men_shoe_transformed, women_shoe_transformed])

# Drop any NaN values
shoe_data_complete_df = shoe_data_complete_df.dropna()

# Create brand id column and fill with unique values based on the brand name column
shoe_data_complete_df["brand_id"] = shoe_data_complete_df.groupby("brand_name").ngroup()
shoe_data_complete_df = shoe_data_complete_df[pd.to_numeric(shoe_data_complete_df["prices_min"], errors = "coerce").notnull()]
shoe_data_complete_df = shoe_data_complete_df[pd.to_numeric(shoe_data_complete_df["prices_max"], errors = "coerce").notnull()]

# Reset, but keep dataframe index
shoe_data_complete_df = shoe_data_complete_df.reset_index(drop = False)
```

```python
# Select the brand_id and brand_name columns from the shoe_data_complete_df and make a new dataframe called brand_df
brand_df = shoe_data_complete_df[["brand_id","brand_name"]]

# Rename brand_id column to id
brand_df = brand_df.rename(columns = {"brand_id" : "id"})

# Drop duplicates
brand_df = brand_df.drop_duplicates()

# reset index
brand_df = brand_df.reset_index(drop = True)
brand_df
```

```python
# Select the id, brand_id, category, prices_min, prices_max column from shoe_data_complete_df and make a new dataframe called shoe_data_df
shoe_data_df = shoe_data_complete_df[["id", "brand_id", "category", "prices_min", "prices_max"]]
# drop duplicates
shoe_data_df = shoe_data_df.drop_duplicates(subset = ["brand_id"])
shoe_data_df
```
